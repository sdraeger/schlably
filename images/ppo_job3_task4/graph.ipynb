```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
makespans = [44.5, 43.24, 45.96, 46.62, 57.88, 54.7, 37.26, 58.38]
models = ["agent_256", "agent_64", "agent_16", "rand", "EDD", "SPT", "MTR", "LTR"]
plt.figure(figsize = (9, 5))
xLocations = np.arange(len(makespans))

bars = plt.bar(xLocations, makespans)
axes = plt.gca()
axes.bar_label(bars)
plt.ylabel("Average Makespan")
plt.xlabel("Scheduler")
plt.title("Average Makespan vs Scheduler")
plt.xticks(xLocations, models)
```

```python
reward = [-3.70833, -3.60333, -3.83, -3.885, -4.82333, -4.55833, -3.105, -4.865]
models = ["agent_256", "agent_64", "agent_16", "rand", "EDD", "SPT", "MTR", "LTR"]
plt.figure(figsize = (9, 5))
xLocations = np.arange(len(reward))

bars = plt.bar(xLocations, reward)
axes = plt.gca()
axes.bar_label(bars)
plt.axhline(0, color = "black", linewidth = 1)  
plt.ylabel("Average Reward")
plt.xlabel("Scheduler")
plt.title("Average Reward vs Scheduler")
plt.xticks(xLocations, models)
```
